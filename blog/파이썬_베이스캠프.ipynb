## ch2. 변수

* 변수는 담는 것이 아니라 가리키는 것. (사물에 이름을 붙이는 포스트잇에 비유)

    * 각 나라의 언어를 변수명으로 사용할 수 있습니다.
    * 대문자로 시작하는 변수는 일반적인 경우 Class로 간주합니다.
    * 스네이크 표기법(파이썬) : hello_world (띄어쓰기를 언더바로 표현)
    * 카멜 표기법(자바스크립트) : helloWorld (띄어쓰기가 있을 자리 뒤의 문자를 대문자로 표현)

````python
s =  'hello world'
s[0:5:2]    # 처음 문자부터 5번째 문자를 2번째마다 건너뛴다
s[0::2]     # stop을 생략하면 자동적으로 마지막까지
s[::2]      # start를 생략하면 자동적으로 처음부터
s[::-1]     # step을 -1로 입력하면 역순으로 출력한다
````

````python
# 문자열 메서드
s = 'hello world'
type(s)
dir(s)

'count'
'find'
'format'
'index'
'isdigit'
'join'
'lower'
'replace'
'split'
'strip'
'upper'
'zfill'
````

````python
# 문자열 메서드
s = 'hello world'

s.count('l')                        # s에서 l이 몇개 있는지 세줘
s.find('world')                     # s에서 wolrd가 몇번째 문자열에 있는지 알려줘
s.index('world')                    # find와 같은 기능이다. 하지만 없는것을 찾아달라고 했을 떄 find는 -1, index는 error가 출력된다.
't1'.isdigit()                      # 문자가 숫자인지 판별
'-'.join(['011','9999','8888'])     # 문자와 합쳐서 배열
'HELLO WORLD'.lower()               # 소문자로 변경
'hellow world'.upper()              # 대문자로 변경
'hello world'.replace('hello','hi') # 앞의 문자를 뒤의 문자로 변경
'hello world'.split(' ')            # '' 안의 기준으로 문자를 쪼개라
'011-9999-8888'.split('-')
'   hello world     '.strip()       # 앞뒤 공백을 없애라
'   hello world     '
'5'.zfill(2)                        # 5를 두자리로 맞춰라

# 'count'
# 'find'
# 'format'
# 'index'
# 'isdigit'
# 'join'
# 'lower'
# 'replace'
# 'split'
# 'strip'
# 'upper'
# 'zfill'
````

## ch.6 논리 자료형과 None 자료형

````python
# 논리 자료형(bool)
x = True    # True : 참을 나타냅니다. 1과 같은 값으로 취급
y = False   # False : 거짓을 나타냅니다. 0과 같은 값으로 취급

type(x)
````

````python
True < 3
````

````python
True == 1
````

````python
False == 0
````

````python
# 다른 언어는 True, true로 표기하는 경우도 있습니다. 따라서 꼭 T를 대문자로 써주세요.
````

````python
bool('hello')   # 문자열이 있는 것은 True로 평가된다.
````

````python
bool('')        # 문자열이 없는 것은 False로 평가된다.
````

````python
print(bool(0))  # 숫자는 0만 False로 평가된다.
print(bool(1))
print(bool(100))
print(bool(-1))
print(bool(-100))
````

````python
print(bool([]))     # 안이 비어있으면 모두 다 False로 평가된다.
print(bool({}))
print(bool(set()))
print(bool(()))
````

````python
# None 자료형 (비어있다, 아무것도 없다, 정보가 없다를 표현하고 싶을 때 사용)

x = None
type(x)
````

````python
None == 0   # 0도 값일 수 있으므로 None은 0이 아니다.
````

````python
None + None # None은 사칙연산을 지원하지 않는다.
````
